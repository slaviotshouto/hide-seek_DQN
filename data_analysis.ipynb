```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.cm as cm

import pandas as pd
from plot_helper import plot_only_mean_rewards, plot_interaction
```

```python
mean_scores_hiders = []
hider_values = []

with open('pandas_plots/hider_rewards.csv', 'r') as f_obj:
    l = f_obj.read().splitlines()
    total_reward = 0
    for idx, value in enumerate(l):
        total_reward += int(value)
        hider_values.append(int(value))
        dev_by = idx+1
        mean_scores_hiders.append(total_reward/dev_by)

mean_scores_seekers = []
seeker_values = []
with open('pandas_plots/seeker_rewards.csv', 'r') as f_obj:
    l = f_obj.read().splitlines()
    total_reward = 0
    for idx, value in enumerate(l):
        total_reward += int(value)
        seeker_values.append(int(value))
        dev_by = idx+1
        mean_scores_seekers.append(total_reward/dev_by)

plot_only_mean_rewards(mean_scores_hiders, mean_scores_seekers)
```

```python
with open('pandas_plots/hiders_interaction.csv', 'r') as f_obj:
    l = f_obj.read().splitlines()
    hider_interaction = []
    for i in l:
        hider_interaction.append(int(i))

with open('pandas_plots/seekers_interaction.csv', 'r') as f_obj:
    l = f_obj.read().splitlines()
    seeker_interaction = []
    for i in l:
        seeker_interaction.append(int(i))
    
with open('pandas_plots/random_hider_rewards.csv', 'r') as f_obj:
    l = f_obj.read().splitlines()
    random_hider_reward = []
    for i in l:
        random_hider_reward.append(int(i))
        
with open('pandas_plots/random_seeker_rewards.csv', 'r') as f_obj:
    l = f_obj.read().splitlines()
    random_seeker_reward = []
    for i in l:
        random_seeker_reward.append(int(i))

with open('pandas_plots/random_hiders_interaction.csv', 'r') as f_obj:
    l = f_obj.read().splitlines()
    random_hider_interaction = []
    for i in l:
        random_hider_interaction.append(int(i))
        
with open('pandas_plots/random_seekers_interaction.csv', 'r') as f_obj:
    l = f_obj.read().splitlines()
    random_seeker_interaction = []
    for i in l:
        random_seeker_interaction.append(int(i))
```

```python
print(len(hider_interaction))
print(len(seeker_interaction))
print(len(hider_values), len(seeker_values))
print(len(mean_scores_hiders),  len(mean_scores_seekers))
```

```python
plot_interaction(hider_interaction, "Hider Item Interaction")
df_hider_interaction = pd.DataFrame(hider_interaction)
df_hider_interaction.describe()
```

```python
plot_interaction(seeker_interaction, "Seeker Item Interaction")
df_seeker_interaction = pd.DataFrame(seeker_interaction)
df_seeker_interaction.describe()
```

```python
plt.clf()
plt.title('Object Interactions')
plt.xlabel('Number of Games')
plt.ylabel('Interactions')    
plt.plot(mean_scores_hiders, label='Hider mean score')
#plt.plot(mean_scores_seekers, label='Seeker mean score')
plt.scatter([i for i in range(1, 2001)], hider_interaction, label='Hider Item Interaction')
plt.legend(loc='upper right')
plt.show(block=False)
```
